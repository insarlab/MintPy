## Read and plot file in SNAP BEAM-DIMAP format

```python
%matplotlib inline

import os
import numpy as np
import matplotlib.pyplot as plt
from mintpy.utils import readfile

proj_dir = os.path.expanduser('~/insarlab/test/SNAP_SAfricaS1/')
os.chdir(proj_dir)
print('Go to directory:', proj_dir)
```

```python
# unwrapped interferogram
fname = os.path.join(proj_dir, '20190415_20190427/20190415_20190427_unw_tc.data/Unw_Phase_ifg_15Apr2019_27Apr2019_VV.img')
data, atr = readfile.read(fname)

# plot
data[data == 0] = np.nan # get rid of zero value, which is filled during phase unwrapping
plt.figure()
plt.imshow(data, cmap='jet')
plt.colorbar()
plt.show()
```

```python
# DEM
fname = os.path.join(proj_dir, 'dem_tc.data/dem_VV.img')
data, atr = readfile.read(fname)

# plot
data[data == -32768] = np.nan # get rid of -32768 value
plt.figure()
plt.imshow(data, cmap='jet')
plt.colorbar()
plt.show()
```
